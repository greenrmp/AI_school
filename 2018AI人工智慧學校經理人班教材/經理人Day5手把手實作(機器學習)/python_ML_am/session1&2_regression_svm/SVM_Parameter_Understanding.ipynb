```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 5
```

```python
from myfun import plot_decision_regions
def svm_example(plot_dict):
    # import some data to play with
    iris = datasets.load_iris()
    X = iris.data[:, :2]  # we only take the first two features.
    y = iris.target
    """
    調整gamma和degree並重新執行看看下面產生的圖形
    提示1. 將gamma調很大看看結果
    提示2. 將degree調成1~3(調超過5以上運算時間會比較久)
    """
    gamma = 0.7
    degree = 1
    C= 1
    # Plot the decision boundary. For that, we will assign a color to each
    for kernel in plot_dict:
        models = svm.SVC(kernel=kernel,#可以選擇linear, poly, rbf 
                         C=C,#調整gamma可以改變rbf kernel的複雜度 
                         degree=degree,#調整degree可以改變poly kernel的複雜度 
                         gamma=gamma)#C越大代表對錯誤的容忍度越大 
        
        models.fit(X,y)
        accuracy = models.score(X,y)
        plt.subplot(plot_dict[kernel])
        plt.tight_layout()
        #plt.figure()
        plot_decision_regions(X, y, models)
        # Plot also the training points
        plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
        plt.xlabel('Sepal length')
        plt.ylabel('Sepal width')
        plt.title('Plot for %s'%kernel + '\n' + 'Accuracy:%.2f'%accuracy)
    plt.show()
```

# Kernel Comparison

```python
plot_dict= {'linear':131, 'poly':132, 'rbf':133}
svm_example(plot_dict)
```
